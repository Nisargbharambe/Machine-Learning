Out of 800, 386 Type 2 variables are missing. So , it is better to drop Type 2 attribute.

```python
df.drop(['Type 2'],inplace=True,axis=1)
```

```python
df.columns
```

```python
df.drop(['Name'],inplace=True,axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
poke_cat = df['Type 1']
poke_cat_encoded = encoder.fit_transform(poke_cat)
print(poke_cat_encoded[0:10])
print('Min Encoded Label  Max Encoded Label')
print(poke_cat_encoded.min(), '               <->          ' , poke_cat_encoded.max())
```

```python
print(encoder.classes_)
df['Type 1'] = poke_cat_encoded
```

```python
print(df.shape)
X = df.iloc[:,0:9]
print(X.shape)
y = df.iloc[:,-1]
print(y.shape)
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.svm import SVC
skfolds = StratifiedKFold(n_splits=2,random_state=42)
model = SVC()
for train_index,test_index in skfolds.split(X,y):
    #print("Train ",train_index," Test",test_index)
    X_train, X_test = X.iloc[train_index] , X.iloc[test_index]
    y_train, y_test = y.iloc[train_index] , y.iloc[test_index]
    model.fit(X_train , y_train)
    y_pred=model.predict(X_test)
    n_correct = sum(y_pred == y_test)
    print(n_correct/len(y_pred))
```
